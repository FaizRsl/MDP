```python
import cv2 
import matplotlib.pyplot as plt

img = cv2.imread('PretrainingDataset/A/images/14_IMG_0870-Medium_jpeg.rf.017e6648b0250e68e5a4b8793e6e78c9.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)

# extract the hue channel from the hsv image
hue = img_hsv[:,:,0]

# increment the hue values by 60
hue = hue + 65

# if pixel values become > 180, subtract 180 
# cond = hue[:, :] > 180
# hue[cond] = hue[cond] - 180
# assign the modified hue channel to hsv image
img_hsv[:,:,0] = hue
im = cv2.cvtColor(img_hsv, cv2.COLOR_HSV2RGB)

plt.imshow(im)
```

```python
import numpy as np

def plotHue(filename):
    img = cv2.imread(filename)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img)
    plt.show()
    img_hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)

    hue = img_hsv[:,:,0]
    plt.title('hue channel')
    plt.imshow(hue)
    plt.show()

    hist, _ = np.histogram(hue, bins=180, normed=True)
    plt.title("hue histogram")
    plt.plot(hist)
    plt.show()


plotHue('PretrainingDataset/A/images/1_A_PNG.rf.50031031daaaeb59a51d8b1676ada560.jpg')
```
