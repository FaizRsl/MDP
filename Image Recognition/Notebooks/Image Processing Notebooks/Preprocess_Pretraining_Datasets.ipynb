```python
import os
import shutil
import yaml
```

```python
# Global Label to ImageID Mapping
labelDict = {
    'A': 20,
    'B': 21,
    'C': 22,
    'D': 23,
    'E': 24,
    'F': 25,
    'G': 26,
    'H': 27,
    'S': 28,
    'T': 29,
    'U': 30,
    'V': 31,
    'W': 32,
    'X': 33,
    'Y': 34,
    'Z': 35,
    'Up': 36,
    'Down': 37,
    'Right': 38,
    'Left': 39,
    'Stop': 40,
    'One': 11,
    'Two': 12,
    'Three': 13,
    'Four': 14,
    'Five': 15,
    'Six': 16,
    'Seven': 17,
    'Eight': 18,
    'Nine': 19,
    'Box': 1,
    'Bullseye': 2,
    'NULL': 3
}
```

```python
destParent = "PretrainingDataset"
#os.mkdir("PretrainingDataset")
```

```python
for name in labelDict.keys():
    os.makedirs(f"{destParent}/{name}", exist_ok=True)
    os.makedirs(f"{destParent}/{name}/images", exist_ok=True)
    os.makedirs(f"{destParent}/{name}/labels", exist_ok=True)
    
os.makedirs(f"{destParent}/Multiple", exist_ok=True)
os.makedirs(f"{destParent}/Multiple/images", exist_ok=True)
os.makedirs(f"{destParent}/Multiple/labels", exist_ok=True)

os.makedirs(f"{destParent}/NULL", exist_ok=True)
os.makedirs(f"{destParent}/NULL/images", exist_ok=True)
os.makedirs(f"{destParent}/NULL/labels", exist_ok=True)
```

```python
def processDataset(parentFolder, datasetMappings):
    with open(f"{parentFolder}/data.yaml", "r") as stream:
        yamlData = yaml.safe_load(stream)
        numClasses = int(yamlData['nc'])
        nameClasses = yamlData['names']

        for childFolder in ['train','valid','test']:
            annotationFolder = f"{parentFolder}/{childFolder}/labels/"
            imgFolder = f"{parentFolder}/{childFolder}/images/"

            annotationFilenames = os.listdir(annotationFolder)
            for annotationFilename in annotationFilenames:
                with open(f"{annotationFolder}/{annotationFilename}") as f:
                    lines = f.readlines()
                    if len(lines) == 0:
                        continue

                    newLines = []
                    for line in lines:
                        line = line.split()
                        label = nameClasses[int(line[0])]
                        
                        if label in labelDict:
                            labelNum = labelDict[label]
                            
                        else:
                            labelNum = labelDict[datasetMappings[label]]
                            label = datasetMappings[label]
                            
                        line[0] = str(labelNum)
                        line = " ".join(line)
                        newLines.append(line)

                    if len(lines) > 1:
                        with open(f"{destParent}/Multiple/labels/{parentFolder}_{annotationFilename}", 'w') as f:
                            for newLine in newLines:
                                f.write(newLine+"\n")
                        shutil.copy(f"{imgFolder}/{annotationFilename[:-4]+'.jpg'}", f"{destParent}/Multiple/images/{parentFolder}_{annotationFilename[:-4]+'.jpg'}") 

                    else:
                        with open(f"{destParent}/{label}/labels/{parentFolder}_{annotationFilename}", 'w') as f:
                            for newLine in newLines:
                                f.write(newLine+"\n")
                        shutil.copy(f"{imgFolder}/{annotationFilename[:-4]+'.jpg'}", f"{destParent}/{label}/images/{parentFolder}_{annotationFilename[:-4]+'.jpg'}") 
```
